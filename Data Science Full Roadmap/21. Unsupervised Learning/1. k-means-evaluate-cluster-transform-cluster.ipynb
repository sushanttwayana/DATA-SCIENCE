```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np

# Load the Iris dataset
iris = load_iris()

# Get the first 10 samples for training
training_points = iris.data[:10]

# Create a KMeans instance with 3 clusters
model = KMeans(n_clusters=3, random_state=42)

# Fit model to the first 10 points
model.fit(training_points)

# Get cluster labels for training points
training_labels = model.labels_

# Create new points for prediction (using samples 10-20)
new_points = iris.data[10:20]
new_labels = model.predict(new_points)

# Plot training points
xs_train = training_points[:, 0]
ys_train = training_points[:, 2]

plt.scatter(xs_train, ys_train, c=training_labels, cmap='viridis', s=100, label='Training data')

# Plot new predictions
xs_new = new_points[:, 0]
ys_new = new_points[:, 2]

plt.scatter(xs_new, ys_new, c=new_labels, cmap='viridis', s=150, marker='X', 
            edgecolor='black', label='New predictions')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('KMeans Clustering (Trained on 10 samples)')
plt.legend()
plt.show()

print("Training labels:", training_labels)
print("New predictions:", new_labels)
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Get the samples
points = iris.data

# Create a KMeans instance with 3 clusters
model = KMeans(n_clusters=3, random_state=42)

# Fit model 
model.fit(points)

# Get cluster labels for all original points
all_labels = model.labels_

# Create completely new random samples within the range of Iris data
np.random.seed(42)
new_points = np.random.uniform(
    low=iris.data.min(axis=0),
    high=iris.data.max(axis=0),
    size=(5, 4)  # 5 new samples with 4 features each
)

# Predict cluster labels for new_points
labels = model.predict(new_points)

print("New sample points:")
print(new_points)
print("\nPredicted cluster labels:")
print(labels)

# Plot ALL data points (original dataset)
xs_all = points[:, 0]
ys_all = points[:, 1]
plt.scatter(xs_all, ys_all, c=all_labels, alpha=0.3, cmap='viridis', label='Original data')

# Plot new points
xs_new = new_points[:, 0]
ys_new = new_points[:, 1]
plt.scatter(xs_new, ys_new, c=labels, alpha=0.8, cmap='viridis', s=100, edgecolor='black', label='New points')

# Plot centroids
centroids = model.cluster_centers_
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]
plt.scatter(centroids_x, centroids_y, marker='D', s=150, c='red', edgecolor='black', label='Centroids')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('KMeans Clustering of Iris Dataset with New Points')
plt.legend()
plt.show()
```

# Trasnforming features for better clusterings

## Scaling fish data for clustering
You are given an array samples giving measurements of fish. Each row represents an individual fish. The measurements, such as weight in grams, length in centimeters, and the percentage ratio of height to length, have very different scales. In order to cluster this data effectively, you'll need to standardize these features first. In this exercise, you'll build a pipeline to standardize and cluster the data.

These fish measurement data were sourced from the Journal of Statistics Education.



* Create an instance of StandardScaler called scaler.
* Create an instance of KMeans with 4 clusters called kmeans.
* Create a pipeline called pipeline that chains scaler and kmeans. To do this, you just need to pass them in as arguments to make_pipeline().

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Random seed for reproducibility
np.random.seed(42)

n_samples = 100  # per species

# --- Species distributions (mean ± std) ---
# Bream: heavy, long
bream_weight = np.random.normal(600, 80, n_samples)
bream_length = np.random.normal(35, 4, n_samples)
bream_ratio = np.random.normal(18, 2, n_samples)

# Pike: long, lighter
pike_weight = np.random.normal(500, 70, n_samples)
pike_length = np.random.normal(40, 5, n_samples)
pike_ratio = np.random.normal(15, 2, n_samples)

# Roach: medium
roach_weight = np.random.normal(300, 50, n_samples)
roach_length = np.random.normal(25, 3, n_samples)
roach_ratio = np.random.normal(22, 3, n_samples)

# Smelt: very light, small
smelt_weight = np.random.normal(100, 20, n_samples)
smelt_length = np.random.normal(15, 2, n_samples)
smelt_ratio = np.random.normal(28, 3, n_samples)

# --- Combine all ---
weights = np.concatenate([bream_weight, pike_weight, roach_weight, smelt_weight])
lengths = np.concatenate([bream_length, pike_length, roach_length, smelt_length])
ratios  = np.concatenate([bream_ratio, pike_ratio, roach_ratio, smelt_ratio])

samples = np.column_stack((weights, lengths, ratios))

species = (["Bream"] * n_samples +
           ["Pike"] * n_samples +
           ["Roach"] * n_samples +
           ["Smelt"] * n_samples)

samples
```

```python
# Perform the necessary imports
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Create scaler: scaler
scaler = StandardScaler()

# Create KMeans instance: kmeans
kmeans = KMeans(n_clusters = 4)

# Create pipeline: pipeline
pipeline = make_pipeline(scaler, kmeans)

```

## Clustering the fish data
You'll now use your standardization and clustering pipeline from the previous exercise to cluster the fish by their measurements, and then create a cross-tabulation to compare the cluster labels with the fish species.

As before, samples is the 2D array of fish measurements. Your pipeline is available as pipeline, and the species of every fish sample is given by the list species.


* Fit the pipeline to the fish measurements samples.
* Obtain the cluster labels for samples by using the .predict() method of pipeline.
* Using pd.DataFrame(), create a DataFrame df with two columns named 'labels' and 'species', using labels and species, respectively, for the column values.
* Using pd.crosstab(), create a cross-tabulation ct of df['labels'] and df['species'].

```python
# Import pandas
import pandas as pd

# Fit the pipeline to samples
pipeline.fit(samples)

# Calculate the cluster labels: labels
labels = pipeline.predict(samples)

# Create a DataFrame with labels and species as columns: df
df = pd.DataFrame({'labels': labels, 'species': species})

# Create crosstab: ct
ct = pd.crosstab(df['labels'], df['species'])

# Display ct
print(ct)
```

Excellent! It looks like the fish data separates really well into 4 clusters!
