```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("/kaggle/input/netflix/NetFlix.csv")

netflix_df.head()
```

```python
# Start coding here! Use as many cells as you like
duration = netflix_df["type"]
print(duration)
```

```python
#  Filter the dataset to include only movies

movies_df = netflix_df[netflix_df["type"] == "Movie"]
print(movies_df)
```

# Filter for movies released between 1990 and 1999

```python
movies_90s = movies_df[(movies_df['release_year'] >= 1990) & (movies_df['release_year'] < 2000)]
print(movies_90s)
```

```python
# Find the most frequent movie duration

print(movies_90s['duration'])

# movies_90s['duration'] = pd.to_numeric(movies_90s['duration'], errors = 'coerce')
print("------------------------")
# print(movies_90s['duration'])

```

```python
# Find the most frequent movie duration in the 1990s
# Ensure the 'duration' column is numeric
movies_90s['duration'] = pd.to_numeric(movies_90s['duration'], errors='coerce')

# Find the most frequent (mode) movie duration
most_frequent_duration = movies_90s['duration'].mode()[0]

# Print the most frequent duration
print("Most Frequent Duration in the 1990s:", most_frequent_duration)
```

# Count the number of short action movies from the 1990s

```python
# Subset for action genre movies
action_movies_90s = movies_90s[movies_90s['genres'] == "Comedies"]

print(action_movies_90s)

short_movie_count = 0


```

```python
# Initialize the counter for short action movies
short_movie_count = 0

# Iterate through the action movies dataframe
for index, row in action_movies_90s.iterrows():
    # Store the duration
    duration = row["duration"]
    
    # Check if the duration is less than 90 minutes
    if pd.notnull(duration) and duration < 90:
        short_movie_count += 1

# Print the result
print("Number of short action movies in the 1990s:", short_movie_count)
```
