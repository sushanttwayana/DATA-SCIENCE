### data/airbnb_last_review.tsv This is a TSV file containing data on Airbnb host names and review dates.

* listing_id: unique identifier of listing
* host_name: name of listing host
* last_review: date when the listing was last reviewed

# Loading the data


```python
# Import necessary packages
import pandas as pd
import numpy as np
```

```python
# Import csv for prices
airbnb_price = pd.read_csv("/kaggle/input/new-york-city-airbnb-market/airbnb_price.csv")

#Import excel file for room types
airbnb_room_type = pd.read_excel('/kaggle/input/new-york-city-airbnb-market/airbnb_room_type.xlsx') 

# Import TSV for review dates
airbnb_last_review = pd.read_csv('/kaggle/input/new-york-city-airbnb-market/airbnb_last_review.csv')
```

# Merging the three DataFrames

```python
# Join the three data frames together into one

listings = pd.merge(airbnb_price, airbnb_room_type, on='listing_id')
# listings = pd.merge(listings, airbnb_last_review, on='listing_id')
listings.head()
```

Determining the earliest and most recent review dates

```python
# Ensure all column names are standardized
airbnb_price.columns = airbnb_price.columns.str.strip().str.lower()
airbnb_room_type.columns = airbnb_room_type.columns.str.strip().str.lower()
airbnb_last_review.columns = airbnb_last_review.columns.str.strip().str.lower()

# Merge the DataFrames
listings = pd.merge(airbnb_price, airbnb_room_type, on='listing_id')

if 'listing_id' in airbnb_last_review.columns:
    listings = pd.merge(listings, airbnb_last_review, on='listing_id')
else:
    print("Column 'listing_id' not found in airbnb_last_review.")

listings.head()
```

# Determining the earliest and most recent review dates

### What are the dates of the earliest and most recent reviews?
### To use a function like max()/min() on last_review date column,it needs to be converted to datetime type

```python
listings['last_date_review'] = pd.to_datetime(listings['last_review'])
listings['last_date_review']
```

```python
first_reviewed = listings['last_date_review'].min()
last_reviewed = listings['last_date_review'].max()

print(first_reviewed, last_reviewed)
```

# Finding how many listings are private rooms


* How many of the listings are private rooms?
* Since there are differences in capitalization, make capitalization consistent

```python
listings['room_type'] = listings['room_type'].str.lower()
private_room_count = listings[listings['room_type'] == 'private room'].shape[0]

print(private_room_count)
```

# Finding the average price of listings'

### What is the average listing price?
### To convert price to numeric, remove " dollars" from each value

```python
listings['price_clean'] = listings['price'].str.replace('dollars', '').astype(float)

avg_price = listings['price_clean'].mean()
avg_price
```

# Creating a DataFrame with the four solution values

```python
review_dates = pd.DataFrame({
    'first_reviewed' : [first_reviewed],
    'last_reviewed' : [last_reviewed],
    'nb_private_rooms' : [private_room_count],
    'avg_price': [round(avg_price, 2)]
})

print(review_dates)
```
