# Using logic with DataFrames

```python
# Select the customer where Age is greater than 40
customer_age_greater_than_40 = credit_records[credit_records.Customer_Age > 40]
print(customer_age_greater_than_40)



```

```python
# Select the customer whose Status is Married
married_customer = credit_records[credit_records.Marital_Status == "Married"]
print(married_customer)
```

```python
# # Select all customer whose Education_Level is Not Graduate
not_graduate = credit_records[credit_records['Education_Level'] != "Graduate"]
print(not_graduate)
```

#  Creating line plots

```python
# From matplotlib, import pyplot under the alias plt
from matplotlib import pyplot as plt

# Plot Officer Deshaun's hours_worked vs. day_of_week
plt.plot(credit_records.Transaction_Category, credit_records.Revenue)

# Display Deshaun's plot
plt.show()
```

```python
from matplotlib import pyplot as plt

# Filter the first 500 records
first_500_records = credit_records.head(500)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(first_500_records['Transaction_Category'], first_500_records['Revenue'], 'o', markersize=4)

# Add labels and title
plt.xlabel('Transaction Category')
plt.ylabel('Revenue')
plt.title('Transaction Category vs Revenue (First 500 Records)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```
